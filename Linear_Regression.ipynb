```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('transformed_data2.csv')
```

```python
#lets see how much Flat Area is skewed
plt.figure(dpi=100)
plt.hist(data['Flat Area (in Sqft)'],color='red', bins=150)
plt.xlabel('frequency')
plt.ylabel('Flat Area')
plt.title('frequency distribution of flat Area')
```

```python
# observation-> as we can see that data is right skewed , lets make it normally distributed
def trans(sample):
    return np.log(sample)
```

```python
data['Transformed_FlatArea']=data['Flat Area (in Sqft)'].map(trans)
```

```python
plt.figure(dpi=100)
plt.hist(data['Transformed_FlatArea'],color='red', bins=150)
plt.xlabel('frequency')
plt.ylabel('Flat Area')
plt.title('frequency distribution of flat Area')
```

```python
data=data.drop(columns='ID')
data
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
Y=data['Sale_Price']
X=scaler.fit_transform(data.drop(columns='Sale_Price'))
X=pd.DataFrame(data=X,columns=data.drop(columns='Sale_Price').columns)
X.head()
```

```python
X.corr()
```

# Multicollinearity

```python
k=X.corr()
collinear_pair=[[i,j]for i in k.columns for j in k.columns if (k.loc[i,j]<abs(0.5))&(i!=j)]
```

```python
collinear_pair
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=X

VIF=pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)

VIF

```

as we can observe we are getting 7 values which have  VIF greater 5 

```python

def remove_multicol(data):
    vif=pd.Series([variance_inflation_factor(data.values,i)for i in range(data.shape[1])],index=data.columns)
    if vif.max() > 5:
        print("deleted ",vif[vif==vif.max()].index[0])
        data=data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print("samller than 5")
        return data
    
```

```python
for i in range(7):
    vif_data=remove_multicol(vif_data)
vif_data.head()
```

```python
# here we run the loop 7 times as their are seven Multicollinearity in independent variable 
# we have to go every possible independent variable and find multicollinearity and remove it
```

# Test and Train the data

```python
X=vif_data# X will be our independent variables after removing multicollinearity
Y=data['Sale_Price']# this is our dependent variables 
```

```python
from sklearn.model_selection import train_test_split # making the train and test data split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)

X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
predictions=lr.predict(X_test)
print('score of the model ',lr.score(X_test,Y_test))
print('predictions of sale prices' ,predictions)
print('test data of sale price  ',Y_test)
```

```python
# observations as score of the model comes out to be 0.84 we can say that the model is desent 
# if we carefully observe the sale price that are predicted are very near to desired result
```

# Residual  and interperating the results

```python
residuals=predictions-Y_test

residuals_table=pd.DataFrame({'residuals':residuals,'predictions':predictions})
residuals_table=residuals_table.sort_values(by='predictions')
```

```python
z=[i for i in range(int(residuals_table['predictions'].max()))]
k=[0 for i in range(int(residuals_table['predictions'].max()))]
```

```python
plt.figure(dpi=100, figsize=(17,7))

plt.scatter(residuals_table['predictions'],residuals_table['residuals'],color='red', s=3)
plt.plot(z,k,color='blue')
plt.xlabel('prediction points')
plt.ylabel('residuals')
plt.title('residuals points')
plt.legend
```

```python
# Observation ====> The residuals are skewed towards 200000,-200000 which means data is normally distributed
# there are some outliers too which means that data needed to be treated 
```

```python
plt.figure(dpi=100)
plt.hist(residuals_table['residuals'],color='blue',bins=200)
plt.xlabel('frequency')
plt.ylabel('residuals')
plt.title('residuals plot')
plt.legend
```

```python
coefficient_table=pd.DataFrame({'independent_variables':X_train.columns,'coefficent':lr.coef_})
coefficient_table=coefficient_table.sort_values(by='coefficent')
coefficient_table
```

```python
plt.figure(figsize=(14,7),dpi=100)
y=coefficient_table['coefficent']
x=coefficient_table['independent_variables']
plt.barh(x,y)
plt.show()
```
